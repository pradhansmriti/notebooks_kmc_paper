```python
import numpy as np
from scipy.stats import exponnorm
import matplotlib.pyplot as plt
mu=[120,120,125,300]
# Define parameters
sigma = [10,10,15,80]
K = [1,2,4,1]
label=['std salt w/o drug','high salt w drug','std salt w drug','low salt w drug']
color_list=['black','green','red','blue']
# Generate x values
x = np.linspace(0, 800, 800)

# Probability density function (PDF)
for i in range(len(mu)):
    pdf = exponnorm.pdf(x, K[i], loc=mu[i], scale=sigma[i])
    plt.plot(x, pdf, label=f'{label[i]}',color=color_list[i])

plt.legend()
# Cumulative distribution function (CDF)
plt.show()
```

```python
random_value_generation=[[] for i in range(4)]

for j in range(4):
    for i in range(1000):
        random_value_generation[j].append(exponnorm.rvs(K[j], loc=mu[j], scale=sigma[j]))
        plt.hist(random_value_generation[j],bins=100)
plt.show()
```
