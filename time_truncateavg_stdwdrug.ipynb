```python
####time average not truncating at 150 NE but for sweep 10**7
df_cutlist=[]
for i, file in enumerate(stdsalt_drug_files[0:10]):
     # Start with fewer files
    try: 
        df = pd.read_csv(file,chunksize=100000)
        for chunk in df:
            df_cut=chunk[chunk['sweep']<=10**7]
            df_cutlist.append(df_cut)
            del df_cut
            del chunk
        del df

    except Exception as e:
        #print(f"Error processing file {file}: {e}")
        continue
```

```python
all_files_data=[]
min_length=[]
for i, file in enumerate(stdsalt_drug_files[:100]):
     # Start with fewer files
    try:
        print(f"Processing file {i+1}: {os.path.basename(file)}")
        df = pd.read_csv(file,chunksize=100000)
        for chunk in df:
            if (chunk['NE'].iloc[-1]>=150):
                cut_index=chunk[chunk['NE']==150]
               
                sweep_value=chunk.loc[cut_index.index[0],'sweep']
               
                min_length.append(sweep_value)
                df_cut=chunk.loc[:cut_index.index[0]-1]
               
                all_files_data.append(df_cut[['sweep','NE']])
                del df_cut
            else:
                continue
            del chunk
        del df

    except Exception as e:
        #print(f"Error processing file {file}: {e}")
        continue
```

```python

df_min=all_files_data[np.argmax(min_length)]

all_files_data_new=[]
min_sweep=np.max(min_length)

for df in all_files_data:
    if(df['sweep'].iloc[-1]<min_sweep):
        df_extended=pd.DataFrame(columns=['sweep','NE'])
        
        df_extended['sweep']=np.arange(round(int(df['sweep'].iloc[-1]),-2),round(int(min_sweep),-2),100)
        df_extended['NE']=150
        df_new=pd.concat((df,df_extended),ignore_index=True)
        all_files_data_new.append(df_new)
        del df_extended
        del df_new
    else:
        all_files_data_new.append(df)
```
